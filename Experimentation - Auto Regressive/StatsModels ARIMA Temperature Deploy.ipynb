```python
# create a differenced series
def difference(dataset, interval=1):
	diff = list()
	for i in range(interval, len(dataset)):
		value = dataset[i] - dataset[i - interval]
		diff.append(value)
	return numpy.array(diff)

# invert differenced value
def inverse_difference(history, yhat, interval=1):
	return yhat + history[-interval]


# Predict
# multi-step out-of-sample forecast
def predict(X, fitted_model, forecast_steps, start = 0, stop = "inf", global_mean = 0, diff_len = 12):
	forecast = fitted_model.forecast(steps=forecast_steps)
	
	# invert the differenced forecast to something usable
	history = [x for x in X]
	original_history = history.copy()
	fixed_predictions = []

	day = 1
	for yhat in forecast:
		# inverted = series_mean + yhat
		inverted = inverse_difference(history, yhat, diff_len)
		
		# inverted = yhat + global_mean
		print('Day %d: %f' % (day, inverted))
		history.append(inverted)
		fixed_predictions.append(inverted)
		day += 1

	plt.figure(figsize=(15, 5))
	plt.plot(numpy.array(history).flatten(), alpha = 1.0, label = "Prediction", color = "orange")
	plt.plot(original_history, alpha = 1.0, label = "Original Data", color = "blue")
	if stop == "inf" or stop < 0:
		plt.xlim(0, len(numpy.array(history)))
	else:
		plt.xlim(start, stop)
	plt.legend()
	plt.show()

	return fixed_predictions
```

```python
def showMonthlyMeanPerYearMultiDF(dfs, titles):
    div = math.ceil(math.sqrt(len(dfs)))

    fig, axs = plt.subplots(div, div, figsize = (24, 7))
    
    for idx, df in enumerate(dfs):
        years = df.groupby(by = "year").groups
        year_keys = list(years.keys())
        vals = df[["datetime","temp", "month", "year", "month_year"]].copy()
        
        curr_x = idx % div 
        curr_y = idx // div 

        for i in range(len(year_keys)):
            curval = vals.loc[years[year_keys[i]], ["datetime", "temp", "month"]].copy()
            val_to_plot = curval.groupby(by = "month").mean(numeric_only = True)["temp"].values
            axs[curr_x, curr_y].plot(val_to_plot, alpha = 0.5, label = year_keys[i])
            axs[curr_x, curr_y].set_title(titles[idx])

        axs[curr_x, curr_y].plot(df.copy().groupby(by = "month").mean(numeric_only = True)["temp"].values, alpha = 1.0, color = "black")
        # axs[curr_x, curr_y].legend()
    
    plt.show()


    return (fig, axs)
    
```

```python
# Acquiring Data
df = raw_data[["datetime", "temp", "tempmin", "tempmax"]].copy()
df["month"] = DatetimeIndex(raw_data["datetime"]).month
df["year"] = DatetimeIndex(raw_data["datetime"]).year 
df["month_year"] = df["year"] * 10000 + df["month"]
```

```python
# Calculate Yearly Means
yearly_means = df.groupby(by = "year").mean(numeric_only = True)["tempmin"]

# Construct Yearly Difference of Means Dataframe
yearly_diff = pd.DataFrame(yearly_means.copy())
yearly_diff["prev_year"] = yearly_diff["tempmin"].shift(1)
yearly_diff.dropna(inplace = True)
yearly_diff["diff"] = yearly_diff["prev_year"] - yearly_diff["tempmin"]
yearly_diff["diff"].plot(figsize = (24, 5))

plt.scatter(yearly_diff.index, yearly_diff["diff"].values)
plt.show()
print("Diff MEAN:", yearly_diff["diff"].mean())
```

```python
year_df = df.copy()

# Remove Yearly Means for each year in the original dataframe
for year in yearly_means.index:
    year_df.loc[year_df["year"] == year, "temp"] -= yearly_means[year]

running_yearly_mean = year_df["temp"].mean()
year_df["temp"] -= running_yearly_mean
```

```python
month_df = df.copy()

# Remove Monthly Means for each month in the dataframe
monthly_means = month_df.groupby(by = "month").mean(numeric_only = True)["temp"]
for month in monthly_means.index:
    month_df.loc[month_df["month"] == month, "temp"] -= monthly_means[month]

running_month_mean = month_df["temp"].mean()

month_df["temp"] -= running_month_mean
```

```python
month_year_df = df.copy()

# Remove Monthly Means for each month in the dataframe
monthly_yearly_means = month_year_df.groupby(by = "month_year").mean(numeric_only = True)["temp"]

for year in yearly_means.index:
    month_year_df.loc[month_df["year"] == year, "temp"] -= yearly_means[year]

month_year_df["temp"] -= running_yearly_mean

for month_year in monthly_yearly_means.index:
    month_year_df.loc[month_df["month_year"] == month_year, "temp"] -= monthly_yearly_means[month_year]

running_month_year_mean = month_year_df["temp"].mean()

month_year_df["temp"] -= running_month_year_mean
```
